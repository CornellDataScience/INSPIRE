```python
import numpy
import pandas
import sklearn
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.decomposition import PCA
```

```python
df = pandas.read_csv("modeless_clustered.csv")
```

```python
arr = numpy.array(df)[:,1:13]
```

```python
df.iloc[:,1:13]
```

```python
dist = euclidean_distances(arr[:,1:], arr[:,1:12])
#for every i, find lowest non-zeros
```

```python
def closest_ten(dist_row):
    return numpy.argsort(dist_row)[1:11]
```

```python
closest_ten(dist[0,:])
```

```python
def song_list_vector(distance_matrix, song_array):
    new_arr = []
    for row in range(len(distance_matrix)):
        ten_indices = closest_ten(distance_matrix[row,:])
        ten_songs = []
        for i in ten_indices:
            ten_songs.append(song_array[i,0])
        ten_songs = numpy.array(ten_songs)
        new_arr.append(ten_songs)
    return numpy.array(new_arr)
    
```

```python
dist.shape
```

```python
temp = numpy.append(arr,song_list_vector(dist, arr), axis = 1)
```

```python
temp = pandas.DataFrame(temp)
```

```python
temp.head()
```

```python
temp.to_csv("songs_with_recommendations.csv")
```

```python
c1 = []
c2 = []
for i in range(len(arr)):
    clust = arr[i, 11]
    if clust == 1:
        c1.append(arr[i,:])
    else:
        c2.append(arr[i,:])
c1 = numpy.array(c1)
c2 = numpy.array(c2)
```

```python
c2
```

```python
clust1 = []
for array in [c1, c2]:
    print(array.shape)
    distances = euclidean_distances(array[:,1:], array[:,1:12])
    clust1.append(song_list_vector(distances, array))
    
```

```python
rec_c1 = numpy.append(c1,clust1[0], axis = 1)
rec_c2 = numpy.append(c2,clust1[1], axis = 1)
```

```python
recs_within_cluster = numpy.append(rec_c1, rec_c2, axis = 0)
```

```python
pca = PCA(n_components = 2)
pca.fit(recs_within_cluster[:,1:11])
transformed_data = pca.fit_transform(recs_within_cluster[:,1:11])
#This is clustered without mode but will need to remove mode from the PCA!! I.e. do this over
```

```python
cluster_recs_w_pca = numpy.append(recs_within_cluster, transformed_data, axis = 1)
```
