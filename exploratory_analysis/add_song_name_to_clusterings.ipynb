```python
import json
import pandas as pd
import numpy as np
import csv
```

```python
with open('song_id_and_names.json') as json_file:  
    data = json.load(json_file)
    
```

```python
print(data['songs'])
```

```python
song_id_to_name_dict = {}
for mapping in data['songs']:
    song_id_to_name_dict[mapping['song_id']] = mapping['song_name']
```

```python
print(song_id_to_name_dict)
```

```python
recommendations_csv = pd.read_csv('../app/static/recommendations_with_components.csv')
```

```python
recommendations_csv["name"] = ""
```

```python
for i, row in recommendations_csv.iterrows():
      recommendations_csv.at[i,'name'] = song_id_to_name_dict[recommendations_csv.at[i, 'id']]
```

```python
print(recommendations_csv)
```

```python
recommendations_csv.to_csv('recommendations_with_names.csv')
```
