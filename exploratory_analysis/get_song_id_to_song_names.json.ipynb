```python
all_songs = {}  
all_songs['songs'] = []

i = 0
for song in song_ids_list:
    track = sp.track(song)
    song_name = track['name']
    artist = track['artists'][0]['name']
    all_songs['songs'].append({"song_id": song, "song_name": song_name, "artist": artist})
    if i % 100 == 0:
        time.sleep(2)
        print("i = ", i)
    i+=1
```
